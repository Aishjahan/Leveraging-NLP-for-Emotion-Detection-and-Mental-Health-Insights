```python
import numpy as np
import matplotlib.pyplot as plt

# Data
models = ['Logistic Regression', 'Naive Bayes', 'ALBERT', 'BERT', 'DistillBERT', 'RoBERTa']
precision = [71.5, 73.0, 83.2, 80.5, 81.0, 78.9]
recall = [72.1, 74.4, 84.5, 82.0, 82.4, 80.0]
f1_score = [71.8, 73.7, 83.8, 81.2, 81.7, 79.4]

# Bar width
bar_width = 0.2
index = np.arange(len(models))

# Light pastel colors
precision_color = '#A8DADC'  # Light blue
recall_color = '#F1C6B1'     # Soft peach
f1_score_color = '#B4E1D1'   # Mint green

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
bar1 = ax.bar(index, precision, bar_width, label='Precision', color=precision_color)
bar2 = ax.bar(index + bar_width, recall, bar_width, label='Recall', color=recall_color)
bar3 = ax.bar(index + 2 * bar_width, f1_score, bar_width, label='F1-Score', color=f1_score_color)

ax.set_xlabel('Models')
ax.set_ylabel('Scores (%)')
ax.set_title('Precision, Recall, and F1-Score Comparison')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(models)
ax.legend()

plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Confusion matrix values (example)
confusion_matrix = np.array([[100, 20, 10],
                             [5, 120, 15],
                             [15, 10, 90]])

# Labels for classes
labels = ['Depression', 'Anxiety', 'PTSD']

# Plot
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix for ALBERT')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Example false positive rate and true positive rate (FPR, TPR)
fpr = [0.0, 0.1, 0.2, 0.3, 0.4, 1.0]
tpr_albert = [0.0, 0.6, 0.75, 0.85, 0.9, 1.0]
tpr_bert = [0.0, 0.5, 0.7, 0.8, 0.85, 1.0]

# Plot ROC Curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr_albert, label='ALBERT (AUC = 0.92)', color='#A8DADC', linewidth=2)  # Light blue
plt.plot(fpr, tpr_bert, label='BERT (AUC = 0.89)', color='#F1C6B1', linewidth=2)      # Soft peach
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Random classifier
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc='lower right')
plt.grid(True, linestyle=':', color='gray', alpha=0.5)
plt.show()
```
